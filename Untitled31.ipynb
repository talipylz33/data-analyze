```python
lr_data = pd.read_csv('D:/RDS_CSV/lr_network_mouse_21122021.rds.csv')

# Mikroglia genlerini adata'dan çıkaralım
microglia_genes = list(adata.var.index)

# Mikroglia genleri için ligand-reseptör eşleşmelerini bul
matching_lr_microglia = lr_data[lr_data['from'].isin(microglia_genes) | lr_data['to'].isin(microglia_genes)]

# İlk birkaç sonucu görelim
print("Mikroglia için Ligand-Reseptör Eşleşmeleri:")
print(matching_lr_microglia.head())
```

```python
# RPE genlerini kullanarak ligand-reseptör eşleşmelerini yapalım
matching_lr_rpe = lr_data[lr_data['from'].isin(rpe_genes_light) | lr_data['to'].isin(rpe_genes_light)]

# İlk birkaç sonucu görelim
print("RPE için Ligand-Reseptör Eşleşmeleri:")
print(matching_lr_rpe.head())
```

```python
# Mikroglia ve RPE genleri arasındaki ligand-reseptör eşleşmelerini bulalım
matching_lr_microglia_rpe = lr_data[lr_data['from'].isin(microglia_genes) & lr_data['to'].isin(rpe_genes_light)]
matching_lr_rpe_microglia = lr_data[lr_data['from'].isin(rpe_genes_light) & lr_data['to'].isin(microglia_genes)]

# Eşleşmeleri kontrol edelim
print("Mikroglia ve RPE Arasındaki Ligand-Receptor Eşleşmeleri (Mikroglia -> RPE):")
print(matching_lr_microglia_rpe.head())

print("\nMikroglia ve RPE Arasındaki Ligand-Receptor Eşleşmeleri (RPE -> Mikroglia):")
print(matching_lr_rpe_microglia.head())
```

```python
import numpy as np
# Etkileşim skorlarını hesaplayacak fonksiyon
def calculate_interaction_score(ligand, receptor):
    # Ligand ve reseptör ekspresyon seviyelerini alalım
    if ligand in adata.var.index and receptor in adata.var.index:
        ligand_expr = np.mean(adata[:, ligand].X)  # Ligand ekspresyonu
        receptor_expr = np.mean(adata[:, receptor].X)  # Reseptör ekspresyonu
        
        # Skor hesaplama (çarpan)
        return ligand_expr * receptor_expr
    else:
        return np.nan  # Eğer genler mevcut değilse

# Mikroglia -> RPE eşleşmeleri için etkileşim skorlarını hesaplayalım
matching_lr_microglia_rpe['Interaction_Score'] = matching_lr_microglia_rpe.apply(
    lambda row: calculate_interaction_score(row['from'], row['to']), axis=1)

# RPE -> Mikroglia eşleşmeleri için etkileşim skorlarını hesaplayalım
matching_lr_rpe_microglia['Interaction_Score'] = matching_lr_rpe_microglia.apply(
    lambda row: calculate_interaction_score(row['from'], row['to']), axis=1)

# Kaynak sütunu ekleyelim
matching_lr_microglia_rpe['source'] = 'Microglia'
matching_lr_rpe_microglia['source'] = 'RPE'

# Sonuçları birleştir
matching_lr_final = pd.concat([matching_lr_microglia_rpe, matching_lr_rpe_microglia])

# Skorları en yüksekten en düşüğe sırala
matching_lr_final_sorted = matching_lr_final.sort_values(by='Interaction_Score', ascending=False)

# İlk 50 eşleşmeyi yazdıralım
print("Mikroglia ve RPE Arasındaki Ligand-Receptor Eşleşmeleri ve Etkileşim Skorları:")
print(matching_lr_final_sorted[['from', 'to', 'source', 'Interaction_Score']].head(50))
```

```python
# Mikroglia -> RPE eşleşmeleri için etkileşim skorlarını hesaplayalım
matching_lr_microglia_rpe['Interaction_Score'] = matching_lr_microglia_rpe.apply(
    lambda row: calculate_interaction_score(row['from'], row['to']), axis=1)

# RPE -> Mikroglia eşleşmeleri için etkileşim skorlarını hesaplayalım
matching_lr_rpe_microglia['Interaction_Score'] = matching_lr_rpe_microglia.apply(
    lambda row: calculate_interaction_score(row['from'], row['to']), axis=1)

# Hem ligand hem de reseptör için source ekleyelim
matching_lr_microglia_rpe['ligand_source'] = 'Microglia'
matching_lr_microglia_rpe['receptor_source'] = 'RPE'

matching_lr_rpe_microglia['ligand_source'] = 'RPE'
matching_lr_rpe_microglia['receptor_source'] = 'Microglia'

# Sonuçları birleştirelim
matching_lr_final = pd.concat([matching_lr_microglia_rpe, matching_lr_rpe_microglia])

# Skorları en yüksekten en düşüğe sırala
matching_lr_final_sorted = matching_lr_final.sort_values(by='Interaction_Score', ascending=False)

# İlk 50 eşleşmeyi yazdıralım
print("Mikroglia ve RPE Arasındaki Ligand-Receptor Eşleşmeleri ve Etkileşim Skorları:")
print(matching_lr_final_sorted[['from', 'to', 'ligand_source', 'receptor_source', 'Interaction_Score']].head(50))
```

```python
import networkx as nx
import plotly.graph_objects as go

# Ağ oluşturma
G = nx.Graph()

# Ligand ve reseptörleri düğümler olarak ekleyelim
for index, row in matching_lr_final_sorted.iterrows():
    ligand = row['from']
    receptor = row['to']
    score = row['Interaction_Score']
    
    # Ligand ve reseptörleri düğüm olarak ekle
    G.add_node(ligand, bipartite=0)  # Ligandlar için part 0
    G.add_node(receptor, bipartite=1)  # Reseptörler için part 1
    
    # Ligand ile reseptör arasındaki kenarı ekleyelim, etkileşim skorunu kenar ağırlığı olarak verelim
    G.add_edge(ligand, receptor, weight=score)

# Ağın düzenini ayarlayalım (bipartite düzenleme kullanarak)
pos = nx.spring_layout(G)

# Kenarların ağırlığını (etkileşim skorunu) kullanarak kenar kalınlıklarını ayarlayalım
edges = G.edges(data=True)
edge_weights = [edge[2]['weight'] for edge in edges]

# Düğümleri ve kenarları Plotly'ye uygun formatta hazırlayalım
edge_x = []
edge_y = []
for start, end in G.edges():
    x0, y0 = pos[start]
    x1, y1 = pos[end]
    edge_x.append(x0)
    edge_x.append(x1)
    edge_y.append(y0)
    edge_y.append(y1)

# Kenar ağırlığını (skor) yansıtan renkler ekleyelim
edge_trace = go.Scatter(
    x=edge_x, y=edge_y,
    line=dict(width=2, color='gray'),
    hoverinfo='none',
    mode='lines')

# Düğümleri yerleştirelim
node_x = []
node_y = []
for node in G.nodes():
    x, y = pos[node]
    node_x.append(x)
    node_y.append(y)

node_trace = go.Scatter(
    x=node_x, y=node_y,
    mode='markers',
    hoverinfo='text',
    marker=dict(
        showscale=True,
        colorscale='YlGnBu',
        size=10,
        colorbar=dict(
            thickness=15,
            title='Node Connections',
            xanchor='left',
            titleside='right'
        )
    ))

# Düğüme metin ekleyelim (ligand ve reseptör isimleri)
node_text = [f"{node}" for node in G.nodes()]
node_trace.marker.color = [len(list(G.neighbors(node))) for node in G.nodes()]

# Şimdi ağı çizelim
fig = go.Figure(data=[edge_trace, node_trace],
                 layout=go.Layout(
                    showlegend=False,
                    hovermode='closest',
                    title="Ligand-Receptor Interaction Network",
                    xaxis=dict(showgrid=False, zeroline=False),
                    yaxis=dict(showgrid=False, zeroline=False),
                    height=700,
                    width=700
                 ))

# HTML dosyasına kaydedelim
fig.write_html('matching_lr_final_sorted.html')

print("Ağ görselleştirmesi 'matching_lr_final_sorted.html' dosyasına kaydedildi.")
```
